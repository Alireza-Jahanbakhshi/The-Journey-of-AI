```python
def get_odds(num_list: list):
    for num in num_list:
        if num % 2 == 0:
            num_list.remove(num)
    return num_list
```

##### Python - 7 - Homework Solution (Game of Life)

The Game of Life, also known simply as Life, is a cellular automaton devised by the British mathematician John Horton Conway in 1970. It is a zero-player game, meaning that its evolution is determined by its initial state, requiring no further input from humans. It is a classic example of emergent behavior; that is, complex patterns emerge from the application of simple rules. It's often used in computer science and mathematical teaching to illustrate how complex patterns can emerge from simple rules, and it's an interesting challenge for programming practice.

Rules of the Game of Life
The universe of the Game of Life is an infinite, two-dimensional orthogonal grid of square cells, each of which is in one of two possible states, alive or dead. Every cell interacts with its eight neighbors, which are the cells that are horizontally, vertically, or diagonally adjacent. At each step in time, the following transitions occur:

Birth: A dead cell with exactly three live neighbors becomes a live cell (reproduction).
Death by Isolation: A live cell with fewer than two live neighbors dies (underpopulation).
Survival: A live cell with two or three live neighbors lives on to the next generation.
Death by Overcrowding: A live cell with more than three live neighbors dies (overpopulation).
The initial pattern constitutes the seed of the system. The first generation is created by applying the above rules simultaneously to every cell in the seed—births and deaths occur simultaneously, and the discrete moment at which this happens is sometimes called a tick.

Objectives for the Homework
Your task is to write a Python program that simulates the Game of Life. Here are the detailed objectives for your homework:

Create a Grid: Implement a way to represent the grid. This could be a 2D list (array) where each element represents a cell and its state (alive or dead).
Initialize the Grid: Allow the user to specify the size of the grid and the initial state of some cells (or create a random initial state).

```python
import pygame
import numpy as np

# Set up dimensions and grid
WIDTH, HEIGHT = 800, 600
ROWS, COLS = 80, 60
CELL_SIZE = WIDTH // COLS, HEIGHT // ROWS

# Initialize the grid randomly
grid = np.random.choice([0, 1], size=(ROWS, COLS))

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Game of Life")

def update_grid(grid):
    new_grid = grid.copy()
    for i in range(1, ROWS-1):
        for j in range(1, COLS-1):
            # Count neighbors
            neighbors = grid[i-1:i+2, j-1:j+2].sum() - grid[i, j]
            
            # Apply rules
            if grid[i, j] == 1 and (neighbors < 2 or neighbors > 3):
                new_grid[i, j] = 0
            elif grid[i, j] == 0 and neighbors == 3:
                new_grid[i, j] = 1
    return new_grid

def draw_grid():
    for i in range(ROWS):
        for j in range(COLS):
            color = (255, 255, 255) if grid[i, j] == 0 else (0, 0, 0)
            pygame.draw.rect(screen, color, (j * CELL_SIZE[0], i * CELL_SIZE[1], CELL_SIZE[0], CELL_SIZE[1]))

# Main loop
running = True
clock = pygame.time.Clock()

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    grid = update_grid(grid)

    screen.fill((255, 255, 255))
    draw_grid()

    pygame.display.flip()
    clock.tick(10)  # Adjust the speed by changing the argument here

pygame.quit()
```
