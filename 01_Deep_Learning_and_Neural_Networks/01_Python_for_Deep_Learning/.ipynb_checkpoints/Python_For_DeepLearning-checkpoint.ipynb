###### Built-in Data Types

In programming, data type is an important concept.

Variables can store data of different types, and different types can do different things.

Python has the following data types built-in by default, in these categories:

Text Type:	str

Numeric Types:	int, float, complex

Sequence Types:	list, tuple, range

Mapping Type:	dict

Set Types:	set, frozenset

Boolean Type:	bool

Binary Types:	bytes, bytearray, memoryview

None Type:	NoneType

Example	                                      Data Type	
x = "Hello World"	                            str	
x = 20	                                        int	
x = 20.5	                                    float	
x = 1j	                                        complex	
x = ["apple", "banana", "cherry"]	            list	
x = ("apple", "banana", "cherry")	            tuple	
x = range(6)	                                range	
x = {"name" : "John", "age" : 36}	            dict	
x = {"apple", "banana", "cherry"}	            set	
x = frozenset({"apple", "banana", "cherry"})    frozenset	
x = True	                                    bool	
x = b"Hello"	                                bytes	
x = bytearray(5)	                            bytearray	
x = memoryview(bytes(5))	                    memoryview	
x = None	                                    NoneType

```python
def get_odds(num_list: list):
    for num in num_list:
        if num % 2 == 0:
            num_list.remove(num)
    return num_list
```

##### Python - 7 - Homework Solution (Game of Life)

```python
import numpy as np
from matplotlib import pyplot as plt
import time
from IPython.display import clear_output
from time import sleep
import random
```

```python
universe = np.zeros((100,100))
for i in [random.randint(40,60) for _ in range(7)]:
    for j in [random.randint(40,60) for _ in range(7)]:
        universe[i, j] = 1
```

```python
plt.imshow(universe)
```

```python
def find_neighbours(cell_ind: tuple, universe: np.array):
    num_neighbours = 0
    l0 = universe.shape[0]
    l1 = universe.shape[1]

    for i in [-1, 0, 1]:
        for j in [-1, 0, 1]:
            if (i!=0) or (j!=0):
                if cell_ind[0] + i == l0:
                    i = - cell_ind[0]
                if cell_ind[1] + j == l1:
                    j = - cell_ind[1]                    
                if universe[cell_ind[0] + i, cell_ind[1] + j] == 1:
                    num_neighbours += 1
    
    return num_neighbours
```

```python
def pass_time(universe):
        num_neighbours_map = np.zeros(universe.shape) 
        
        for i in range(universe.shape[0]):
            for i in range(universe.shape[1]):
                num_neighbours_map[i, j] = find_neighbours((i,j), universe)
                
        new_universe = np.zeros(universe.shape) 
        survivors = (universe == 1) & ((num_neighbours_map == 2) | (num_neighbours_map == 3))
        zombies = (universe == 0) & (num_neighbours_map ==3)
        new_universe = (survivors == 1) | (zombies == 1)
        
        return new_universe
```

```python
num_turns = 50
new_universe = universe
for _ in range(num_turns):
    new_universe = pass_time(new_universe)
    plt.imshow(new_universe, cmap='Greys')
    plt.show()
    sleep(1)
    clear_output(wait=True)
```

```python
def make_pizza(size, topping_0, topping_1, topping_2):
    '''
    Print the summary of the ordered pizza and return the number
    of tippings.
    '''
    print(f"{size}-inch Pizza was ordered with:")
    print(f" -{topping_0}")
    print(f" -{topping_1}")
    print(f" -{topping_2}")
    
make_pizza(8, 'Beef', 'Onion', 'Mushroom')
```

```python
def make_pizza(size, toppings_list):
    '''
    Print the summary of the ordered pizza and return the number
    of tippings.
    '''
    print(f"{size}-inch Pizza was ordered with:")
    for topping in toppings_list:
        print(f" -{topping}")
    
make_pizza(8, ['Beef', 'Onion', 'Mushroom'])
```

```python
#Decorating a function
#Decorators
def my_decorator(func):
    
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
        
    return wrapper

def say_hi():
    print("Hi")
    
say_hi_2 = my_decorator (say_hi)
```
